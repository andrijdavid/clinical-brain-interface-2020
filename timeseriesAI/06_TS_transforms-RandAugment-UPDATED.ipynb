# Time Series Data Augmentation & RandAugment

It's a know fact that a way to improve performance of DL models is to get more labeled data. However, this is usually expensive, or impossible. 

A good alternative to getting more labeled data is data augmentation. Data augmentation techniques create multiple synthetic samples that maintain the same label as the original ones.

There are plenty of data augmentation techniques that can be applied to images. However, there are not so many that can be applied to time series. In this notebook, you'll find several examples of these time series data augmentation techniques that you'll be able to apply to your datasets.

## Time series transforms 

I've added a number of new single-item transforms to the fastai_timeseries library that are now available for use. 
In general there are 2 types of transforms.

1) Geometric transforms: those that slightly modify the time series in the x and/ or y-axes 
- TSmagnoise == TSjittering (1)
- TSmagscale (1)
- TSdimmagscale (1)
- TSmagwarp (1)
- TStimenoise (1)
- TStimewarp (1)
- TSzoomin (3)
- TSrandtimestep
- TSzoomout (3)
- TSrandomzoom


2) Erasing transforms: those that hide/delete certain part of the time series, setting the values to zero or just removing them:
- TSlookback 
- TStimestepout (1) * modifies the output shape
- TStimestepzero (1)
- TSdimout = TSchannelout
- TScutout (2)
- TScrop (1)
- TSrandomcrop (3)
- TScentercrop (3)
- TSmaskout


All of them can be used independently or in combination. In this section, you'll see how these transforms work.

In all tfms, magnitude determines the amount of deviation from the original ts.

‚ò£Ô∏è ***All these transforms can be applied on a single ts (use as a regular tfm) or as a train_dl tfms (applying it to a batch) which is much faster.***

Data augmentation of wearable sensor data for parkinson's disease monitoring using convolutional neural networks. arXiv preprint arXiv:1706.00527. (includes: Jittering, Scaling, Magnitude-Warping, Time-warping, Random Sampling, among others.

(2) Inspired by DeVries, T., & Taylor, G. W. (2017). 
Improved regularization of convolutional neural networks with cutout. arXiv preprint arXiv:1708.04552.

## RandAugment

This has been presented in a new paper in Sep 2019: 
Cubuk, E. D., Zoph, B., Shlens, J., & Le, Q. V. (2019). 
RandAugment: Practical data augmentation with no separate search. arXiv preprint arXiv:1909.13719.

It is a new method of applying data augmentation that has proven to be very useful across several datasets, setting SOTA performance records.

The main idea is that RandAugment will randomly select N (usually between 1-3) augmentations randomly selected from a list. A random augmentation will thus be applied to every single batch. 

There are 2 parameters that can be tuned:

- N: number of tfms to be applied to each individual batch. Default is 1. Usually between 1-3.
- magnitude: if not passed, each tfms will apply its default value. If a value is passed (usually between .05 and .3) this is passed to the tfms.

To use RandAugment, you need to pass it a list of tfms. 

For convenience, I have pre-configured 4 different lists: 

- TS_geometric_tfms TS_xy_tfms
- TS_erasing_tfms == TS_zero_tfms
- TS_tfms (this is usually a good starting point!)
- all_TS_tfms (TS_tfms + TSlookahead + TScentercrop (which are special tfms)

```python
#export

def TS_geometric_tfms(**kwargs):
    return [[
        TStimewarp(**kwargs),
        TSmagwarp(**kwargs),
        TStimenoise(**kwargs),
        TSmagnoise(**kwargs),
        TSmagscale(**kwargs),
        TSdimmagscale(**kwargs),
        TSzoomin(**kwargs),
        TSzoomout(**kwargs),
        TSrandomzoom(**kwargs),
        TSrandtimestep(**kwargs),
    ], []]


def TS_erasing_tfms(**kwargs):
    return [[
        
        TSdimout(**kwargs),
        TScutout(**kwargs),
        TStimestepzero(**kwargs),
        TScrop(**kwargs),
        TSrandomcrop(**kwargs),
        TSmaskout(**kwargs)
    ], []]


def TS_tfms(**kwargs):
    return [TS_geometric_tfms(**kwargs)[0] + TS_erasing_tfms(**kwargs)[0], []]


def all_TS_tfms(**kwargs):
    return [TS_tfms(**kwargs)[0] + 
            [TStimestepout(**kwargs), TSlookback(**kwargs), TScentercrop(**kwargs)], 
            []]
```

```python
tfms = RandAugment(TS_tfms(), N=1)
tfms(xb).shape
```

```python
tfms = RandAugment(TS_tfms(), N=2, magnitude=.1)
tfms(xb).shape
```

```python
%timeit RandAugment(TS_tfms(), N=1)(ts)
```

```python
for i in range(100): plt.plot(RandAugment(TS_tfms(), N=1)(ts).T, color='gainsboro')
plt.plot(ts.T)
plt.show()
```

### RandAugment

With RandAugment there are only 2 parameters you may want to tune: 

- N: number of tfms applied to each batch
- magnitude: how strong should the transformation be

Here are a few examples of how you can do this, instead of testing so many tfms, or combinations of them.

```python
db.train_ds.tfms = None # Just to make sure there are no other tfms passed to the databunch
```

```python
model = ResNet(db.features, db.c).to(device)
learn = Learner(db, model,metrics=accuracy).randaugment()
learn.fit_one_cycle(50)
learn.recorder.plot_losses()
learn.recorder.plot_metrics()
```

```python
model = ResNet(db.features, db.c).to(device)
learn = Learner(db, model,metrics=accuracy).randaugment(N=2)
learn.fit_one_cycle(50)
learn.recorder.plot_losses()
learn.recorder.plot_metrics()
```

```python
model = ResNet(db.features, db.c).to(device)
learn = Learner(db, model,metrics=accuracy).randaugment(N=3)
learn.fit_one_cycle(50)
learn.recorder.plot_losses()
learn.recorder.plot_metrics()
```

```python
model = ResNet(db.features, db.c).to(device)
learn = Learner(db, model,metrics=accuracy).randaugment(N=2, magnitude=.2)
learn.fit_one_cycle(50)
learn.recorder.plot_losses()
learn.recorder.plot_metrics()
```

In this case, it seems that N=2 or N=3 may be good values. With them you actually achieve the best outcomes without a complex search, as the search space is very reduced (usually just test N=1, 2, or 3, as the magnitude default values tend to work well).
